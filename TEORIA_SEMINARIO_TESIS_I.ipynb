# **1.INTRODUCCION**
---


Hoy en dia, tenemos herramientas imprescindibles que dependen del uso de posicionamiento satelital.

- Guiar un avión automaticamente.
- Procesos cotidianos como usar el celular para guiarnos por la calle.
- Aumento de la navegación autónoma.
- Comunicaciones satélitas, las cuales son un factor relevante en la vida de las personas.
- Acividades cotidianas como realizar conversaciones hasta las mas sofisticadas y complejas conexiones satélitales para enviar informacion encriptada de vital importancia entre embajadas o de seguridad nacional.


Estas tecnologías pueden verse afectadas por diversos fenómenos físicos ionosféricos que actualmente no terminan de ser modelados como el fenómeno de la **F-dispera** y el **centello ionosférico o cintilaciones**.

La **ionósfera** es la **region** del espacio donde operan diversas misiones satelitales y es el medio por donde se propagan las señales de **telecomunicaciones y de radionavegacion**. Se denomina **ionósfera**  a la región del espacio que se encuentra entre los **60 a 2000 km de altura** aproximadamente y eta compuesto de plasma o gas ionizado que interatua con **campos eléctricos y magnético**

Las señales de radiofrecuencia empleadas en los sistemas de
navegación y telecomunicaciones pueden sufrir perturbaciones al atravesar estructuras de plasma o irregularidades en la ionósfera, comprometiendo el correcto funcionamiento de estos sistemas debido a la degradación o interrupción de las señales.

Entre estas **irregularidades ionoféricas** se encuentran las burbujas de plasma ecuatoriales que son regiones espaciales en la alta atmósfera donde se produce un decremento irregular de la densidad del plasma respecto a la densidad que las rodea y a las que se conoce como el **fenómeno de la F-Dispersa**.**Relacionado a la aparición de burbujas de plasma** durante las horas de la noche en las bajas latitudes, se observa el **CENTELLEO IONOSFÉRICO O CINTILACIONES**.

Las CINTILACIONES se define como las perturbaciones de amplitud y fase de las señales de radio que se generan cuando atraviesan las burbujas de plasma. Es oportuno y preciso señalar que estos fenómenos forman parte del clima espacial, el cual es determinado por la interacciónde los procesos fisicos entre **el sol y la tierra**.

El clima espacial se define como las condiciones fisicas del:
- Sol
- El medio interplanetario
- El campo magnético
- La atmosféra terrestre
- La superficie terrestre que se encuetran influenciados por la interaccion Sol-Tierra

Las variaciones en el **Clima Espacial** pueden provocar serias afectaciones en nuestra tecnología y por eso de alta relevancia la investigacion de sus efectos.

El **CENTELLEO** es más frecuente en latitudes bajas y altas, es asi que las latitudes medias, como en los **EEUU**, experimentan el centelleo con mucha menos frecuencia que en las bajas latitudes como en el Peru, donde es mucho mas intenso. Ademas, es importante mencionar que el centello es un funcion importante de:
- Hora local.
- Estación del año.
- Acividad geomagnética
- Ciclo Solar

Ante el aumento en la demanda en los sistemas en los cuales el posicionamiento se vuelve un factor critico de seguridad(safety-critical-applications) debido al alto crecimiento en el uso de la industria de vehiculos autonomos, el monitoreo de la Ionósfera y el uso de técnias de mitigación de efectos ionosféricos se vuelve mas importante brindar respaldo a estos sistemas.

Con relación a trabajos sobre pronósticos de fenómenos en la alta atmósfera, previamente se han realizado estudios relacinados a la predicción de **cintilaciones**.**Un primer precedente** es el trabajo **FIRST(Forecasting Scintillation Activity and Equatorial Spread F)[8]**, en este trabajo se utiliza la altura virtual de la capa F, la cual es comparada con un threshold para la misma. **Este threshold** se determina en funcion de la actividad solar. De esta manera en funcion al Threshold se establece la posible actividad de Cintilaciones en 3 niveles.

El **segundo trabajo es Forecasting scintillation activity and equatorial spread F[9]**, el cual establece como variable a analizar el parámetro del drift de plasma ExB, en donde se establece una velocidad de Drifts para la prediccion S4, ademas en esta investigacion se asume que la no presencia de F dispersa y la no presencia de S4 de nivel alto también esta **correlacionada.**

Además existen investigaciones que hacen uso de Machine Learning, el primer método que se revisó fue visto en **Ionospheric Scintillation Forcasting model based on NN-PSO technique[10]**, en este paper se usa una red neuronal para predecir S4 hasta 4 horas entre  los parámetros utilizados para la prediccion destacan el f10.7, kp index, y h'f que es la altura virtual de la capa F.


En el estudio **"New Capabilities for prediction of High-Latitude ionospheric scintillation: a novel approach with machine learning"[11]** se realiza una prediccion con 1 hora de anticipación utilizando el metodo de Support Vector Machine. La misma aproximacion utilizada en esta investigación fue continuada en **[12]**, la diferenia es que en este caso se utilizó una red neuronal y agregaron variables de Kp y f10.7 previas a modo de memoria, en este caso la predicción fue de una hora.

El objetivo de esta etapa es realizar el desarrollo de un sistema de pronóstico de **Cintilaciones** para que sirva de alerta ante la ocurrencia de estos eventos que afectana a las aplicaciones que utilizan señales Satélitales. En este proyecto se va a utilizar:

- **INDICE S4**: Este parámetro nos brinda la medida de Centelleo en la Amplitud de las señales de Radio. Se define como la **raiz cuadrada de la varianza normalizada de la Intensidad de la señal de radiofrecuencia**  sobre un intervalo dado, este intervalo de tiempo es del orden de 1 minuto.
- De manera equivalente, el **INDICE S4** es la desviación estandar sobre el promedio de la intensidad de la señal(eq. 1).

---

$$S_{4}(I)=\sqrt{\frac{<I^{2}>- <I>^{2}}{<I>^{2}}}$$

---



* $$S_{4}: Indice-de-Centelleo-de-Amplitud$$
* $$I : Intensidad-de-la-señal$$

### **CICLOS SOLARES**

Los **ciclos solares** son períodos de actividad del Sol que se caracterizan por variaciones en la cantidad de manchas solares, radiación solar, y fenómenos como las erupciones solares y eyecciones de masa coronal. Cada ciclo solar dura aproximadamente **11 años en promedio**, aunque puede variar entre 9 y 14 años.

Características principales de los ciclos solares:
1. Manchas solares:

Son regiones más frías y oscuras en la superficie del Sol causadas por intensos campos magnéticos.
La cantidad de manchas solares aumenta y disminuye durante el ciclo solar.
El número de manchas solares alcanza un máximo solar y luego disminuye hacia un mínimo solar.
2. Fases del ciclo solar:

Mínimo solar:
Período de baja actividad, con pocas manchas solares.
Menor emisión de radiación ultravioleta y partículas cargadas.
Máximo solar:
Período de alta actividad, con muchas manchas solares y frecuentes erupciones solares.
Mayor emisión de radiación y actividad geomagnética.

3. Campo magnético solar:

Durante cada ciclo, los polos magnéticos del Sol se invierten.
Este cambio de polaridad ocurre cada 11 años aproximadamente, lo que hace que un ciclo completo de polaridad (de vuelta a su estado original) tome unos 22 años.

4. Efectos en la Tierra:

Auroras boreales y australes: Aumentan durante el máximo solar debido a la interacción del viento solar con la magnetosfera terrestre.
Clima espacial: La actividad solar afecta las comunicaciones, los satélites, los sistemas de navegación y la ionosfera.
Variaciones en la radiación cósmica: En el máximo solar, la radiación cósmica que llega a la Tierra disminuye debido a la mayor actividad del viento solar, que actúa como una barrera protectora.

Representación del ciclo solar:
Se representa generalmente como un gráfico conocido como el número de manchas solares de Wolf, que muestra cómo varía el número de manchas solares a lo largo del tiempo.

Relación con la ionosfera:
Los ciclos solares afectan directamente a la ionosfera terrestre, especialmente durante el máximo solar, cuando el aumento en la radiación ultravioleta y rayos X intensifica la ionización, afectando señales GPS, telecomunicaciones, y causando fenómenos como las cintilaciones ionosféricas.

Ciclo solar actual (Ciclo 25):
El ciclo solar actual, conocido como el Ciclo Solar 25, comenzó en diciembre de 2019. Los pronósticos indican que alcanzará su máximo solar entre 2025 y 2026.
