## **¿Que es, cuando y como ocurre el Centelleo Ionosférico?**

- Irregularidades en la densidad de electrones suele correlacionarse con perturbaciones del plasma ionosférico
- Estas variaciones que provocan fluctuaciones, en respuesta generan **Centelleo Ionosférico**, las cuales suelen ocurrir con frecuencia en regiones delatitud bajas.
- Este artículo describe la combinación de una ANN con el algoritmo genético(GA) para predecir **Centelleo Ionosférico**. El algoritmo genetico(GA) se utiliza para obtener los pesos inciales del modelo de ANN.

- Este fenómeno suele ocurrir en el rango de latitud de 15 a 20 grados del ecuador magnético, lo que se llama anomalía ionosférica ecuatorial (EIA), y suele tener una densidad electrónica mayor que la de otras regiones.

Las burbujas de plasma ionosférico se generan inicialmente (**generalmente después de la puesta del sol**) en La parte inferior de la región F alrededor del ecuador magnético, que se elevan a mayores altitudes y se extienden a lo largo de las líneas del campo magnético hasta latitudes bajas.
- Este aumento está relacionado con un proceso inestable llamado Rayleigh Taylor, que conduce a perturbaciones en la ionosfera, El plasma empobrecido, que contribuye al proceso de burbujeo, se combina con el plasma condensado y sin reaccionar producido en la ionosfera ecuatorial, creando un fuerte gradiente en la estructura ionosférica. Este proceso produce graves perturbaciones en la región y comienza a aumentar el centelleo ionosférico en las regiones ecuatoriales de la ionosfera.


## **Consecuencias del Centelleo Ionosférico**

- El **centelleo ionosférico de amplitud** puede reducir el nivel de señal transmitida , bloquear el umbral o causar desvanecimientos profundos de la señal de satélite en el caso de centelleo ionosferico intenso.
- Este fenómeno provoca una desviacion de la señal en direcciones aletaorias alrededor del eje principal de propagacion de la onda creando perturbaciones en el indice de refraccion del medio.
- El efecto del **centelleo ionosférico sobre la fase portadora** provoca deslizamientos de ciclo, desconecta la señal del receptor y provoca pérdidas de cerradura.
- El centelleo ionosferico de fase puede crear un cambio de frecuencia en la señal y aumentar la cantidad de ancho de banda, lo que bloquea la fase de receptor y reduce la precisión de la capacidad del receptor para conectarse y fijarse en la señal transmitida desde el satélite. En consecuencia, este proceso aumenta la precision de la dilución geométrica del parametro de precion y por lo tanto aumenta la **tasa de error al determinar la posicion del receptor** reduciendo en consecuenciala capacidad del receptor para recibir y  rastrear la señales transmitidas.


## **¿Porque modelar y predecir el centelleo ionosférico?**

* Reducir los efectos de su error en las observaciones de GNSS y por tanto aumentar la precision de la aplciacion de comunicacion y navegación por satelite.
* Hasta el momento se han propuesto diferentes modelos empiricos para estimar la cantidad de amplitud y fase del centelleo ionosférico.
* La predicción de este fenómeno mediante modelos matemáticos es de gran importancia ya que estos métodos pueden usarse para modelar estos fenómenos mediante parámetros empiricos.
* Aunque algunos modelos empiricos han utilizado parámetros como latitud y longitud, estaciones , y el tiempo de observacion como entrada, el fenómeno de centelleo ionosférico se ve afectado por parámetros fisicos de la ionósfera , además de los parámetros espaciales.

## **3.Physical Parameters Affecting Ionospheric Scintillation**

En esta investigacion los parámetros que afectan a la ionósfera y que puedan causar centelleo ionosférico, se utilizan como entradas GANN. Para estudiar el centelleo ionosférico, diferentes parámetros como:

* La altura maxima de la capa F2(En km)
* La velocidad de deriva vertical(km/s)
* Flujo solar F10.7
* Número de manchas solares
* Indice geomagnético  internacional kp.

El comportamiento de los fenómenos ionosféricos a lo largo del año se puede divirdir en tres estaciones, es decir:
- Solsticio de verano.
- Solsticiode invierno
- Equinoccion

Los meses se pueden clasificar como sigue:
* Enero, Febrero,noviembre y diciembre como meses de solsticio de invierno.
* Mayo, junio,julio y agosto como meses de solsticio de verano.
* Marzo,abril, septiembre y octubre como meses de equinoccio.

Por este motivo seleccionar un mes de cada una de las estaciones y realizar modelizaciones y predicciones para un periodo de treina dias.
En este paper,se utilizaron los meses de enero, marzo y junio del 2017 como datos de entrada para el estudio inicial de GANN.
Entre estos parámetros, los dos parámetros de la altura máxima y la velocidad de deriva vertical de la capa F2 afectan directamente a la capa F2, que es la capa principal del fenómeno de centelleo ionosférico.
